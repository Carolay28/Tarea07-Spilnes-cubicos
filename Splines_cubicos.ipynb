#### EJERCICIO 3

Dir√≠jase al pseudoc√≥digo del spline c√∫bico con frontera natural provisto en clase, en
base a ese pseudoc√≥digo complete la siguiente funci√≥n:

```python
import sympy as sym

def cubic_spline(xs: list[float], ys: list[float]) -> list[sym.Symbol]:
    """
    Cubic spline interpolation ``S``. Every two points are interpolated by a cubic polynomial
    ``S_j`` of the form ``S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3.``

    xs must be different but not necessarily ordered nor equally spaced.

    ## Parameters
    - xs, ys: points to be interpolated

    ## Return
    - List of symbolic expressions for the cubic spline interpolation.
    """

    points = sorted(zip(xs, ys), key=lambda x: x[0])  # sort points by x

    xs = [x for x, _ in points]
    ys = [y for _, y in points]

    n = len(points) - 1  # number of splines

    h = [xs[i + 1] - xs[i] for i in range(n)]  # distances between  contiguous xs

    # Completado
    alpha = [0] * (n + 1)
    
    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [1]
    u = [0]
    z = [0]

    for i in range(1, n):
        l += [2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]]
        u += [h[i] / l[i]]
        
        # Completado
        z.append((alpha[i] - h[i - 1] * z[i - 1]) / l[i])

    l.append(1)
    z.append(0)
    c = [0] * (n + 1)

    x = sym.Symbol("x")
    splines = []
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        
        # Completado
        a = ys[j]
        
        # Completado
        S = a + b * (x - xs[j]) + c[j] * (x - xs[j])**2 + d * (x - xs[j])**3

        splines.append(S)
    splines.reverse()
    return splines
```

#### EJERCICIO 4

Usando la funci√≥n anterior, encuentre el spline c√∫bico para:

$xs = [1, 2, 3]$

$ys = [2, 3, 5]$

```python
import sympy as sym
from typing import List

def cubic_spline(xs: List[float], ys: List[float]) -> List[sym.Expr]:
    points = sorted(zip(xs, ys), key=lambda x: x[0])
    xs = [float(x) for x, _ in points]
    ys = [float(y) for _, y in points]

    n = len(points) - 1
    h = [xs[i + 1] - xs[i] for i in range(n)]

    alpha = [0.0] * (n + 1)
    for i in range(1, n):
        alpha[i] = 3.0 / h[i] * (ys[i + 1] - ys[i]) - 3.0 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [1.0]
    u = [0.0]
    z = [0.0]

    for i in range(1, n):
        l_i = 2.0 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]
        l.append(l_i)
        u.append(h[i] / l[i])
        z.append((alpha[i] - h[i - 1] * z[i - 1]) / l[i])

    l.append(1.0)
    z.append(0.0)
    c = [0.0] * (n + 1)

    x = sym.Symbol("x")
    splines = []
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2.0 * c[j]) / 3.0
        d = (c[j + 1] - c[j]) / (3.0 * h[j])
        a = ys[j]
        S = sym.simplify(a + b*(x - xs[j]) + c[j]*(x - xs[j])**2 + d*(x - xs[j])**3)
        splines.append(S)

    splines.reverse()
    return splines

def imprimir_splines(splines, xs):
    x = sym.Symbol("x")
    for i, S in enumerate(splines):
        print(f"\nSpline S{i}(x) en [{xs[i]}, {xs[i+1]}]:")
        print("Forma sin expandir:")
        print(" S{}(x) = {}".format(i, sym.simplify(S)))

        print("Forma expandida:")
        print(" S{}(x) = {}".format(i, sym.expand(S)))

        print("-" * 40)

xs = [1, 2, 3]
ys = [2, 3, 5]

splines = cubic_spline(xs, ys)
imprimir_splines(splines, xs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cubic_spline_clamped(xs, ys, fp0, fpn):
    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)
    n = len(xs)

    h = np.diff(xs)
    al = np.zeros(n)
    al[0] = 3*(ys[1] - ys[0])/h[0] - 3*fp0
    al[-1] = 3*fpn - 3*(ys[-1] - ys[-2])/h[-1]

    for i in range(1, n-1):
        al[i] = (3/h[i])*(ys[i+1] - ys[i]) - (3/h[i-1])*(ys[i] - ys[i-1])

    l = np.ones(n)
    mu = np.zeros(n)
    z = np.zeros(n)

    l[0] = 2*h[0]
    mu[0] = 0.5
    z[0] = al[0]/l[0]

    for i in range(1, n-1):
        l[i] = 2*(xs[i+1] - xs[i-1]) - h[i-1]*mu[i-1]
        mu[i] = h[i]/l[i]
        z[i] = (al[i] - h[i-1]*z[i-1])/l[i]

    l[-1] = h[-1]*(2 - mu[-2])
    z[-1] = (al[-1] - h[-2]*z[-2])/l[-1]

    c = np.zeros(n)
    b = np.zeros(n-1)
    d = np.zeros(n-1)
    a = ys.copy()

    c[-1] = z[-1]
    for j in range(n-2, -1, -1):
        c[j] = z[j] - mu[j]*c[j+1]
        b[j] = (a[j+1] - a[j])/h[j] - h[j]*(c[j+1] + 2*c[j])/3
        d[j] = (c[j+1] - c[j])/(3*h[j])

    def S(x):
        x = np.array(x)
        result = np.zeros_like(x, dtype=float)
        for i in range(n-1):
            idx = np.where((x >= xs[i]) & (x <= xs[i+1]))
            dx = x[idx] - xs[i]
            result[idx] = a[i] + b[i]*dx + c[i]*dx**2 + d[i]*dx**3
        return result

    return S

# --- Curva 1 ---
x1 = [1,2,5,6,7,8,10,13,17]
y1 = [3.0,3.7,3.9,4.2,5.7,6.6,7.1,6.7,4.5]
fp0_1 = 1.0
fpn_1 = -0.67

# --- Curva 2 ---
x2 = [17,20,23,24,25,27,27.7]
y2 = [4.5,7.0,6.1,5.6,5.8,5.2,4.1]
fp0_2 = 3.0
fpn_2 = -4.0

# --- Curva 3 ---
x3 = [27.7,28,29,30]
y3 = [4.1,4.3,4.1,3.0]
fp0_3 = 0.33
fpn_3 = -1.5

S1 = cubic_spline_clamped(x1, y1, fp0_1, fpn_1)
S2 = cubic_spline_clamped(x2, y2, fp0_2, fpn_2)
S3 = cubic_spline_clamped(x3, y3, fp0_3, fpn_3)

plt.figure(figsize=(13,6))

# Curva 1
xx = np.linspace(min(x1), max(x1), 400)
plt.plot(xx, S1(xx), label="Curva 1")
plt.scatter(x1, y1, color='blue')

# Curva 2
xx = np.linspace(min(x2), max(x2), 400)
plt.plot(xx, S2(xx), label="Curva 2")
plt.scatter(x2, y2, color='orange')

# Curva 3
xx = np.linspace(min(x3), max(x3), 400)
plt.plot(xx, S3(xx), label="Curva 3")
plt.scatter(x3, y3, color='green')

plt.title("Splines c√∫bicos - Curva 1, 2, 3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
